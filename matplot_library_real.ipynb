# Matplot Grid

```python
import matplotlib.pyplot as man
import numpy as khan

y=khan.array([2,5,1,6,7,3,2,1,4,3,2,1,5,6,7,8,9])
x=khan.array([4,2,5,1,7,4,3,1,4,5,6,7,8,9,0,1,2])
man.plot(y)
man.plot(x)

font1={
    'family':'serif',
    'color':'hotpink',
    'size':15}
font2={
    'family':'serif',
    'color':'darkred',
    'size':12}

man.title("Simple-Graph",fontdict=font1)
man.ylabel('Y-axis',fontdict=font1)
man.xlabel('X-axis',fontdict=font2)
man.grid(linewidth=1)
man.show()
```

# subplot

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([2,1,3,4,5,6,7])
plt.subplot(5,1,1)
plt.plot(y)
plt.grid()

y1=np.array([2,1,3,4,5,6,7])
plt.subplot(5,1,2)
plt.plot(y1)
plt.grid()

y3=np.array([2,1,3,4,5,6,7])
plt.subplot(5,1,3)
plt.plot(y3)
plt.grid()

y4=np.array([2,1,3,4,5,6,7])
plt.subplot(5,1,4)
plt.plot(y4)
plt.grid()

y5=np.array([2,1,3,4,5,6,7])
plt.subplot(5,1,5)
plt.plot(y5)
plt.suptitle("simple")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as  plt
import numpy as np

x=np.array([10,5,8,9])
a_label=('day1','day2','day3','day4')
ex=(0.3,0.2,0.1,0.4)

plt.pie(x,labels=a_label,shadow=True,explode=ex)
plt.show()
```

```python
import matplotlib.pyplot as  plt
import numpy as np

x=np.array([10,5,8,9])
a_label=('day1','day2','day3','day4')
ex=(0.3,0.2,0.1,0.4)
c=('pink','red','navy','green')

plt.pie(x,labels=a_label,shadow=True,explode=ex,colors=c,startangle=90)
plt.title("four days")
plt.legend()
plt.show()
```
